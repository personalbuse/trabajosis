Trabajo guiado: dataset de 5 clases (caballo, pollo, vaca, oveja, elefante) usando CNN, Data Augmentation y Transfer Learning. Ejecutar en GPU (Colab/MPS) si es posible. Cada sección incluye celdas Markdown para explicar decisiones y capturar resultados para el informe final.

## Guía rápida de ejecución

2. Ajusta `max_per_class` y fracciones de split para balancear (recomendado recortar a tamaño simétrico por clase).

```python
# Imports básicos
import os, shutil, random, json, math, itertools
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print(tf.__version__)
tf.config.experimental.list_physical_devices()
```

```python
# Configuración general
BASE_DIR = Path().resolve()
RAW_DIR = BASE_DIR / "Animals5"
SPLIT_DIR = BASE_DIR / "data" / "Animals5_split"
train_frac, val_frac, test_frac = 0.5, 0.25, 0.25  # mitad para train, mitad para val+test
max_per_class = None  # opcional: recorta para balancear; usa entero si quieres limitar
image_size = (160, 160)  # reducir para acelerar
batch_size = 32
seed = 42
tf.random.set_seed(seed)
np.random.seed(seed)
```

## Preparación y balanceo de datos
Generamos particiones simétricas por clase (`train`, `val`, `test`). Si `max_per_class` es `None`, usa el mínimo disponible entre clases; si se define, recorta a ese máximo. La copia se hace a `data/Animals5_split/<split>/<clase>` para no modificar el dataset original.

```python
def count_images_per_class(root: Path):
    data = {}
    for cls_dir in sorted(root.glob("*/")):
        if cls_dir.is_dir():
            imgs = [p for p in cls_dir.glob("*.*") if p.is_file()]
            data[cls_dir.name] = len(imgs)
    return data

counts_raw = count_images_per_class(RAW_DIR)
counts_raw
```

```python
def prepare_splits(raw_dir: Path, split_dir: Path, train_frac=0.5, val_frac=0.25, test_frac=0.25, max_per_class=None, seed=42):
    assert abs(train_frac + val_frac + test_frac - 1.0) < 1e-6, "Las fracciones deben sumar 1"
    rng = random.Random(seed)
    split_dir.mkdir(parents=True, exist_ok=True)
    counts = count_images_per_class(raw_dir)
    target_per_class = min(counts.values()) if max_per_class is None else min(max_per_class, min(counts.values()))
    split_counts = {
        "train": math.floor(target_per_class * train_frac),
        "val": math.floor(target_per_class * val_frac),
        "test": target_per_class - math.floor(target_per_class * train_frac) - math.floor(target_per_class * val_frac)
    }
    print(f"Usando {target_per_class} imágenes por clase => {split_counts}")
    for split in ["train", "val", "test"]:
        (split_dir / split).mkdir(parents=True, exist_ok=True)
    for cls, total in counts.items():
        imgs = [p for p in (raw_dir / cls).glob("*.*") if p.is_file()]
        rng.shuffle(imgs)
        imgs = imgs[:target_per_class]
        boundaries = {
            "train": split_counts["train"],
            "val": split_counts["train"] + split_counts["val"]
        }
        split_map = {
            "train": imgs[:boundaries["train"]],
            "val": imgs[boundaries["train"]:boundaries["val"]],
            "test": imgs[boundaries["val"]:]
        }
        for split, subset in split_map.items():
            out_cls = split_dir / split / cls
            out_cls.mkdir(parents=True, exist_ok=True)
            for src in subset:
                dst = out_cls / src.name
                if not dst.exists():
                    shutil.copy2(src, dst)
    return split_counts

split_info = prepare_splits(RAW_DIR, SPLIT_DIR, train_frac, val_frac, test_frac, max_per_class, seed)
counts_split_train = count_images_per_class(SPLIT_DIR / "train")
counts_split_val = count_images_per_class(SPLIT_DIR / "val")
counts_split_test = count_images_per_class(SPLIT_DIR / "test")
counts_split_train, counts_split_val, counts_split_test
```

## Carga con `tf.data`
Utilizamos `image_dataset_from_directory` y aplicamos `cache`/`prefetch` para acelerar. Ajusta `image_size` y `batch_size` según memoria disponible. Se reserva un subconjunto de validación explícito (ya separado en disco).

```python
def make_datasets(split_dir: Path, image_size=(160, 160), batch_size=32, seed=42):
    train_ds = keras.utils.image_dataset_from_directory(
        split_dir / "train",
        image_size=image_size,
        batch_size=batch_size,
        shuffle=True,
        seed=seed)
    val_ds = keras.utils.image_dataset_from_directory(
        split_dir / "val",
        image_size=image_size,
        batch_size=batch_size,
        shuffle=True,
        seed=seed)
    test_ds = keras.utils.image_dataset_from_directory(
        split_dir / "test",
        image_size=image_size,
        batch_size=batch_size,
        shuffle=False)
    class_names = train_ds.class_names
    autotune = tf.data.AUTOTUNE
    train_ds = train_ds.cache().prefetch(buffer_size=autotune)
    val_ds = val_ds.cache().prefetch(buffer_size=autotune)
    test_ds = test_ds.cache().prefetch(buffer_size=autotune)
    return train_ds, val_ds, test_ds, class_names

train_ds, val_ds, test_ds, class_names = make_datasets(SPLIT_DIR, image_size, batch_size, seed)
class_names
```

```python
# Muestra de batch
def show_batch(ds, class_names):
    plt.figure(figsize=(10, 10))
    for images, labels in ds.take(1):
        for i in range(9):
            ax = plt.subplot(3, 3, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            plt.title(class_names[labels[i]])
            plt.axis("off")
    plt.show()

show_batch(train_ds, class_names)
```

## Modelo 1: CNN básica (Conv2D + MaxPool)
Red convolucional de complejidad media. Se optimiza con Adam y `early stopping`. Ajusta `epochs` según tiempo disponible.

```python
def build_cnn(input_shape, num_classes):
    inputs = keras.Input(shape=input_shape)
    x = layers.Rescaling(1./255)(inputs)
    x = layers.Conv2D(32, 3, activation="relu", padding="same")(x)
    x = layers.MaxPool2D()(x)
    x = layers.Conv2D(64, 3, activation="relu", padding="same")(x)
    x = layers.MaxPool2D()(x)
    x = layers.Conv2D(128, 3, activation="relu", padding="same")(x)
    x = layers.MaxPool2D()(x)
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation="relu")(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = keras.Model(inputs, outputs, name="cnn_basica")
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

input_shape = image_size + (3,)
num_classes = len(class_names)
cnn_basica = build_cnn(input_shape, num_classes)
cnn_basica.summary()
```

```python
epochs = 20
early_stop = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True, monitor="val_accuracy")
history_cnn = cnn_basica.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs,
    callbacks=[early_stop]
 )
test_loss, test_acc = cnn_basica.evaluate(test_ds)
print(f"Test acc (CNN básica): {test_acc:.3f}")
```

## Modelo 2: CNN con BatchNormalization + Dropout
Se agrega normalización por lotes para estabilizar el entrenamiento y `Dropout` para reducir sobreajuste.

```python
def build_cnn_bn_dropout(input_shape, num_classes, dropout_rate=0.4):
    inputs = keras.Input(shape=input_shape)
    x = layers.Rescaling(1./255)(inputs)
    for filters in [32, 64, 128]:
        x = layers.Conv2D(filters, 3, padding="same", activation="relu")(x)
        x = layers.BatchNormalization()(x)
        x = layers.MaxPool2D()(x)
    x = layers.Flatten()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(dropout_rate)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = keras.Model(inputs, outputs, name="cnn_bn_dropout")
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

cnn_bn = build_cnn_bn_dropout(input_shape, num_classes)
cnn_bn.summary()

history_cnn_bn = cnn_bn.fit(
    train_ds, validation_data=val_ds, epochs=epochs, callbacks=[early_stop]
)
test_loss_bn, test_acc_bn = cnn_bn.evaluate(test_ds)
print(f"Test acc (CNN BN+Dropout): {test_acc_bn:.3f}")
```

## Modelo 3: CNN con Data Augmentation
Se añade un pipeline de augmentación en el modelo (rotaciones, flips y zoom) para mejorar generalización.

```python
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="data_augmentation")

def build_cnn_aug(input_shape, num_classes, dropout_rate=0.3):
    inputs = keras.Input(shape=input_shape)
    x = data_augmentation(inputs)
    x = layers.Rescaling(1./255)(x)
    for filters in [32, 64, 128]:
        x = layers.Conv2D(filters, 3, padding="same", activation="relu")(x)
        x = layers.MaxPool2D()(x)
    x = layers.Flatten()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(dropout_rate)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = keras.Model(inputs, outputs, name="cnn_aug")
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

cnn_aug = build_cnn_aug(input_shape, num_classes)
cnn_aug.summary()

history_cnn_aug = cnn_aug.fit(
    train_ds, validation_data=val_ds, epochs=epochs, callbacks=[early_stop]
)
test_loss_aug, test_acc_aug = cnn_aug.evaluate(test_ds)
print(f"Test acc (CNN + Aug): {test_acc_aug:.3f}")
```

## Transfer Learning (2 modelos): Feature Extraction y Fine Tuning
Usaremos dos backbones preentrenados en ImageNet: `MobileNetV2` y `EfficientNetB0`.
- Feature extraction: congelar el backbone, entrenar solo el clasificador.
- Fine tuning: descongelar capas finales del backbone para afinar. Ajusta `unfreeze_at` según recursos.
Nota: puede requerir más tiempo y memoria; reduce `image_size` o `batch_size` si es necesario.

```python
def build_transfer_model(base_name: str, input_shape, num_classes, trainable=False, unfreeze_at=None):
    if base_name == "MobileNetV2":
        base = keras.applications.MobileNetV2(
            input_shape=input_shape, include_top=False, weights="imagenet", pooling="avg")
        preprocess = keras.applications.mobilenet_v2.preprocess_input
    elif base_name == "EfficientNetB0":
        base = keras.applications.EfficientNetB0(
            input_shape=input_shape, include_top=False, weights="imagenet", pooling="avg")
        preprocess = keras.applications.efficientnet.preprocess_input
    else:
        raise ValueError("Modelo no soportado")
    base.trainable = trainable
    inputs = keras.Input(shape=input_shape)
    x = data_augmentation(inputs)
    x = preprocess(x)
    x = base(x, training=False)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = keras.Model(inputs, outputs, name=f"{base_name}_{'ft' if trainable else 'fe'}")
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-4 if trainable else 1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    if trainable and unfreeze_at is not None:
        for layer in base.layers[:unfreeze_at]:
            layer.trainable = False
        for layer in base.layers[unfreeze_at:]:
            layer.trainable = True
    return model

transfer_results = []
for base_name in ["MobileNetV2", "EfficientNetB0"]:
    fe_model = build_transfer_model(base_name, input_shape, num_classes, trainable=False)
    fe_history = fe_model.fit(train_ds, validation_data=val_ds, epochs=10, callbacks=[early_stop])
    fe_test_loss, fe_test_acc = fe_model.evaluate(test_ds)
    transfer_results.append({"modelo": base_name, "modo": "feature_extraction", "test_acc": fe_test_acc})
    print(f"{base_name} FE acc: {fe_test_acc:.3f}")
    ft_model = build_transfer_model(base_name, input_shape, num_classes, trainable=True, unfreeze_at=-30)
    ft_history = ft_model.fit(train_ds, validation_data=val_ds, epochs=10, callbacks=[early_stop])
    ft_test_loss, ft_test_acc = ft_model.evaluate(test_ds)
    transfer_results.append({"modelo": base_name, "modo": "fine_tuning", "test_acc": ft_test_acc})
    print(f"{base_name} FT acc: {ft_test_acc:.3f}")
```

## Métricas, matriz de confusión y tabla comparativa
Funciones auxiliares para calcular métricas por modelo y resumir resultados. Guarda estas salidas para el informe final.

```python
from sklearn.metrics import confusion_matrix, classification_report
def eval_model(model, ds, class_names, name):
    y_true, y_pred = [], []
    for batch_imgs, batch_labels in ds:
        preds = model.predict(batch_imgs, verbose=0)
        y_true.extend(batch_labels.numpy())
        y_pred.extend(np.argmax(preds, axis=1))
    cm = confusion_matrix(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    acc = np.mean(np.array(y_true) == np.array(y_pred))
    print(f"{name} accuracy: {acc:.3f}")
    return cm, report, acc

def plot_confusion(cm, class_names, title):
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
    plt.xlabel("Predicción")
    plt.ylabel("Real")
    plt.title(title)
    plt.show()

results = []
models_for_eval = [
    (cnn_basica, "CNN básica"),
    (cnn_bn, "CNN BN+Dropout"),
    (cnn_aug, "CNN + Aug")
]
for entry in transfer_results:
    # Evalúa de nuevo usando diccionario de modelos; almacena accuracy ya calculada.
    results.append({"modelo": entry["modelo"], "modo": entry["modo"], "test_acc": entry["test_acc"]})
for model, name in models_for_eval:
    cm, report, acc = eval_model(model, test_ds, class_names, name)
    plot_confusion(cm, class_names, title=name)
    results.append({"modelo": name, "modo": "scratch", "test_acc": acc})

pd.DataFrame(results)
```
